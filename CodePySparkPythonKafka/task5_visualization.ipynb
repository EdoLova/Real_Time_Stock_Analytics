```python
import matplotlib.pyplot as plt

# Data from the table
events = list(range(1, 16))
entry_price = 70.336
prices = [70.336, 75.858, 73.358, 93.725, 102.350, 112.710,
          133.600, 97.840, 117.760, 110.230, 115.920, 141.020,
          163.890, 65.385, 64.546]

# Computing boundaries TP and SL
take_profit_level = entry_price * 1.20
stop_loss_level   = entry_price * (1 - 0.05)

plt.figure(figsize=(12,6))
plt.plot(events, prices, marker='o', label='AAPL Price', color='blue', linewidth=2)

plt.axhline(y=take_profit_level, color='green', linestyle='--', linewidth=2, label='Take-Profit (20%)')
plt.axhline(y=stop_loss_level, color='red', linestyle='--', linewidth=2, label='Stop-Loss (5%)')

# Higlight datapoints for TP and SL
for i, price in enumerate(prices):
    if price >= take_profit_level:
        plt.scatter(events[i], price, color='green', s=100, zorder=5)
    elif price <= stop_loss_level:
        plt.scatter(events[i], price, color='red', s=100, zorder=5)

plt.xlabel('Event Number', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.xticks(events)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
